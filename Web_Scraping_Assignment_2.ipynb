##### Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
def data_analyst(url):
    driver =  webdriver.Chrome(r"F:\sanss_data\FlipRobo\chromedriver.exe")
    driver.get(url)
    # Creating the emply list for the required data to be scraped.
    job_titles = []
    job_location = []
    company_name  = []
    experience_required = []
    # Scraping the required data.
    for i in driver.find_elements_by_xpath("//a[@class = 'title fw500 ellipsis']")[0:10]:
        job_titles.append(i.text)
    for i in driver.find_elements_by_xpath("//a[@class = 'subTitle ellipsis fleft']")[0:10]:
        company_name.append(i.text)
    for i in driver.find_elements_by_xpath("//li[@class = 'fleft grey-text br2 placeHolderLi location']/span")[0:10]:
        job_location.append(i.text)
    for i in driver.find_elements_by_xpath("//li[@class = 'fleft grey-text br2 placeHolderLi experience']/span")[0:10]:
        experience_required.append(i.text)
    # Creating the dataframe.
    df_data = pd.DataFrame({"Job Title" : job_titles,
                            "Company" : company_name,
                            "Location" : job_location,
                            "Experience Required" : experience_required})
    print(df_data)
    
# calling the function
data_analyst("https://www.naukri.com/data-analyst-jobs-in-bangalore?k=data%20analyst&l=bangalore")
```

###### Q5: Write a python program to scrape the salary data for Data Scientist designation in Noida location. You have to scrape Company name, Number of salaries, Average salary, Min salary, Max Salary.

```python
def salary(url):
    driver = webdriver.Chrome(r"F:\sanss_data\FlipRobo\chromedriver.exe")
    driver.get(url)
    # Creating the emply list for the required data to be scraped.
    company_name = []
    no_of_slaries = []
    avg_salary = []
    min_salary = []
    max_salary = []
    # Scraping the required data.
    for i in driver.find_elements_by_xpath("//p[@class= 'm-0']")[1:30:3]:
        company_name.append(i.text)
    for i in driver.find_elements_by_xpath("//p[@class = 'css-1uyte9r css-1kuy7z7 m-0']")[0:10]:
        no_of_slaries.append(i.text)
    for i in driver.find_elements_by_xpath("//div[@class = 'col-2 d-none d-md-flex flex-row justify-content-end']")[0:10]:
        avg_salary.append(i.text.replace('\n',''))
    for i in driver.find_elements_by_xpath("//div[@class = 'common__RangeBarStyle__values common__flex__justifySpaceBetween common__flex__container ']/span")[0:20:2]:
        min_salary.append(i.text)
    for i in driver.find_elements_by_xpath("//div[@class = 'common__RangeBarStyle__values common__flex__justifySpaceBetween common__flex__container ']/span")[1:20:2]:
        max_salary.append(i.text)
    # Creating the dataframe.
    sal_df = pd.DataFrame({"Company Name" : company_name,
                           "Number of Salaries" : no_of_slaries,
                           "Average Salary" : avg_salary,
                           "Minimum Salary" : min_salary,
                           "Maximum Salary" : max_salary})
    print(sal_df)
    
    
    
# calling the function

salary("https://www.glassdoor.co.in/Salaries/new-delhi-data-scientist-salary-SRCH_IL.0,9_IM1083_KO10,24.htm?clickSource=searchBtn")
```

###### Q6 : Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes: 1. Brand, 2. Product Description, 3. Price, 4. Discount %

```python
def sun_flipkart(url):
    driver = webdriver.Chrome(r"F:\sanss_data\FlipRobo\chromedriver.exe")
    driver.get(url)
    # Creating the emply list for the required data to be scraped.
    brand_name = []
    prod_description = []
    prod_price = []
    prod_discount = []
    # CLosing the login pop-up which incase poped-up after opening the url.
    cross = driver.find_element_by_xpath("//button[@class = '_2KpZ6l _2doB4z']")
    if cross is not None:
        cross.click()
        driver.find_element_by_xpath("//input[@title = 'Search for products, brands and more']").send_keys("Sunglasses")
        driver.find_element_by_xpath("//button[@type = 'submit']").click()
    else:
        driver.find_element_by_xpath("//input[@title = 'Search for products, brands and more']").send_keys("Sunglasses")
        driver.find_element_by_xpath("//button[@type = 'submit']").click()
    pages = np.arange(0,3,1)
    # Loop for iterating each page.
    for page in pages:
        page = "https://www.flipkart.com/search?q=sunglasses&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(page)
        driver =  webdriver.Chrome(r"F:\sanss_data\FlipRobo\chromedriver.exe")
        driver.get(page)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        # Scraping the required data.
        for i in driver.find_elements_by_xpath("//div[@class = '_2WkVRV']"):
            brand_name.append(i.text)
        for i in driver.find_elements_by_xpath("//a[@class = 'IRpwTa _2-ICcC' or @class = 'IRpwTa']"):
            prod_description.append(i.text)
        for i in driver.find_elements_by_xpath("//div[@class = '_30jeq3']"):
            prod_price.append(i.text)
        for i in driver.find_elements_by_xpath("//div[@class = '_3Ay6Sb']/span"):
            prod_discount.append(i.text)
    # Creating the dataframe.
    sunglasses_df = pd.DataFrame({"Brand" : brand_name[0:100],
                                  "Product Description" : prod_description[0:100],
                                  "Price" : prod_price[0:100],
                                  "Discount %" : prod_discount[0:100]})                            
    print(sunglasses_df)
    
# Calling the function
sun_flipkart('https://www.flipkart.com/')
```

##### Q9: Go to the link - https://www.myntra.com/shoes. Set Price filter to “Rs. 6649 to Rs. 13099” , Color filter to “Black”,  then scrape First 100 shoes data you get. The data should include “Brand” of the shoes , Short Shoe description, price of the shoe.

```python
def myntra(url):
    driver = webdriver.Chrome(r"F:\sanss_data\FlipRobo\chromedriver.exe")
    driver.get(url)
    # Creating the emply list for the required data to be scraped.
    brand = []
    short_description = []
    price = []    
    pages = np.arange(0,3,1)
    # Loop for each page.
    for page in pages:
        page = "https://www.myntra.com/shoes?f=Color%3ABlack_36454f&p="+str(page)+"&plaEnabled=false&rf=Price%3A6687.0_13125.0_6687.0%20TO%2013125.0"
        driver =  webdriver.Chrome(r"F:\sanss_data\FlipRobo\chromedriver.exe")
        driver.get(page)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        # Scraping the required data.
        for i in driver.find_elements_by_xpath("//h3[@class = 'product-brand']"):
            brand.append(i.text)
        for i in driver.find_elements_by_xpath("//h4[@class = 'product-product']"):
            short_description.append(i.text)
        for i in driver.find_elements_by_xpath("//div[@class = 'product-price']"):
            price.append(i.text)
    # Creating the dataframe.
    sneakers_df = pd.DataFrame({"Brand" : brand[0:100],
                                "Product Description" : short_description[0:100],
                                "Price" : price[0:100]})
    print(sneakers_df)
    
# Calling the function
myntra('https://www.myntra.com/shoes')
```

##### Q10: Go to webpage https://www.amazon.in/,  Enter “Laptop” in the search field and then click the search icon.
