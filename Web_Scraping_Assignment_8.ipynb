##### 1. Scrape the details of most viewed videos on YouTube from Wikipedia.

```python
# Function Definition
def youtube(url):
    driver = webdriver.Chrome(r"F:\sanss_data\FlipRobo\chromedriver.exe")
    driver.get(url)
    
    # Creating the empty lists for  the data to be extracted.
    rank = []
    name = []
    artist = []
    upload_date = []
    views = []
    list_of_data = [rank, name, artist, upload_date, views]
    
    # Extracting the required data.
    for x in range(1,6):
        for i in driver.find_elements_by_xpath("//tbody/tr/td[" + str(x) + "]")[:30]:
            list_of_data[x-1].append(i.text)
    
    # striping the extra data from name.
    duplicate_name = []
    for n in list_of_data[1]:
        duplicate_name.append(n[:-4])
        
    list_of_data[1] = duplicate_name
    
    # Creating the dataframe.
    most_watched  =  pd.DataFrame({"Rank" : list_of_data[0],
                                  "Name" : list_of_data[1],
                                  "Artist" : list_of_data[2],
                                  "Upload Date" : list_of_data[3],
                                  "Views" : list_of_data[4]})
    return most_watched
    
    
# calling function
youtube("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

##### 3. Scrape the details of selenium exception from guru99.com.

```python
# Function Definition
def guru(url):
    driver = webdriver.Chrome(r"F:\sanss_data\FlipRobo\chromedriver.exe")
    driver.get(url)
    
    # Reaching to the selenium exceptions handling page.
    driver.find_element_by_xpath("//a[@title = 'Selenium']").click()
    driver.find_element_by_xpath("//a[@title = 'Selenium Exception Handling (Common Exceptions List)']").click()
    
    # Creating the empty lists for  the data to be extracted.
    name = []
    description = []
    data_list = [name, description]
    
    # Extrating required data
    for x in range(1,3):
        for i in driver.find_elements_by_xpath("//table[@class = 'table table-striped']/tbody/tr/td[" + str(x) + "]")[1:42]:
            data_list[x-1].append(i.text)
            
    # Creating dataframe.
    exception_df = pd.DataFrame({"Name" : data_list[0],
                            "Description" : data_list[1]})
    return exception_df


# Calling function
guru("https://www.guru99.com/")
```

##### 4. Scrape the details of State-wise GDP of India from statisticstime.com.

```python
# Function Definition
def economy(url):
    driver = webdriver.Chrome(r"F:\sanss_data\FlipRobo\chromedriver.exe")
    driver.get(url)
    
    # Creating the empty lists for  the data to be extracted.
    rank = []
    state = []
    GSDP_18_19 = []
    GSDP_17_18 = []
    share = []
    GDP = []
    
    list_of_data = [GSDP_18_19, share, GDP]
    
    # Reaching to the economy option.
    driver.find_element_by_xpath("//div[@class = 'navbar']/div[2]/button").click()
    
    # Reaching India page.+9999
    try:
        india = driver.find_element_by_xpath("//div[@class = 'navbar']/div[2]/div/a[2]")
        india.click()
    except ElementNotInteractableException as e:
        driver.get(india.get_attribute('href'))
        
    # Reaching to the state wise GDP page.
    driver.find_element_by_xpath("//ul[@style='list-style-type:none;margin-left:20px;']/li/a").click()
    
    # Extracting required data.
    # Rank
    for i in driver.find_elements_by_xpath("//table[@id = 'table_id']/tbody/tr/td[@class = 'data1']"):
        rank.append(i.text)
    # State
    for i in driver.find_elements_by_xpath("//table[@id = 'table_id']/tbody/tr/td[@class = 'name']"):
        state.append(i.text)
    # GSDP_17_18
    for i in driver.find_elements_by_xpath("//table[@id = 'table_id']/tbody/tr/td[@class = 'data sorting_1']"):
        GSDP_17_18.append(i.text)
    # GSDP_18_19, share, GDP
    for x in range(1,4):
        for i in driver.find_elements_by_xpath("//table[@id = 'table_id']/tbody/tr/td[@class = 'data'][" + str(x) + "]"):
            list_of_data[x-1].append(i.text)
            
    # Creating dataframe
    df_GDP  =  pd.DataFrame({"Rank" : rank,
                                  "State" : state,
                                  "GSDP (18-19)" : list_of_data[0],
                                  "GSDP (17-18)" : GSDP_17_18,
                                  "share" : list_of_data[1],
                                  "GDP" : list_of_data[2]})
    return df_GDP



# Calling Function
economy("http://statisticstimes.com/")    
```

##### 5. Scrape the details of trending repositories on Github.com.

```python
# Function Defination
def github(url):
    ch_options = Options()
    ch_options.add_argument('--incognito')
    driver = webdriver.Chrome(r"F:\sanss_data\FlipRobo\chromedriver.exe", options = ch_options)
    driver.get(url)
    
    # Selecting Explore
    time.sleep(2)
    driver.find_element_by_xpath("//nav[@class = 'mt-0 px-3 px-lg-0 mb-5 mb-lg-0']/ul/li[4]").click()
    
    # Reaching to the trending page from dropdown
    time.sleep(2)
    try:
        trend = driver.find_element_by_xpath("//ul[@class = 'list-style-none mb-3'][2]/li[3]/a")
        trend.click()
    except ElementNotInteractableException:
        driver.get(trend.get_attribute('href'))
    
    # Creating empty lists for the data to be extracted.
    rep_title = []
    rep_description = []
    contri_count = []
    lang_used = []
    
    # Extracting the required data.
    # list of article
    all_items = driver.find_elements_by_xpath("//article[@class = 'Box-row']")
    # Repository Title
    for i in driver.find_elements_by_xpath("//article[@class = 'Box-row']/h1/a"):
        rep_title.append(i.text)
    # Contributors Count
    for i in driver.find_elements_by_xpath("//div[@class = 'f6 text-gray mt-2']/a[2]"):
        contri_count.append(i.text)
    # Repository Description and Language Used
    for i in all_items:
        try:
            rep_description.append(i.find_element_by_tag_name('p').text)
        except NoSuchElementException:
            rep_description.append("-") 
    
    # Creating the dataFrame
    trending_df = pd.DataFrame({"Repository Title" : rep_title,
                               "Repository Description" : rep_description,
                               "Contributors Count" : contri_count})
    return trending_df



# Calling Function
github("https://github.com/")
```

##### 6. Scrape the details of top 100 songs on billboard.com.

```python
# Function Definition
def billboard(url):
    driver = webdriver.Chrome(r"F:\sanss_data\FlipRobo\chromedriver.exe")
    driver.get(url)
    
    # Creating the empty lists for the data to be extracted.
    song_name = []
    artist_name = []
    last_week_rank = []
    peak_rank = []
    weeks_on_board = []
    
    # Reaching the charts option
    try:
        charts = driver.find_element_by_xpath("//ul[@class = 'header__link-list header__main-menu bg--dark']/li[1]/a")
        charts.click()
    except ElementNotInteractableException as e:
        driver.get(charts.get_attribute('href'))
    
    # Reaching the Hot 100 page.
    driver.find_element_by_xpath("//div[@class = 'site-header-links__contents bb-user-anon bb-user-free']/a[3]").click()
    
    # Extrating the required data
    # Song Name
    for i in driver.find_elements_by_xpath("//span[@class = 'chart-element__information__song text--truncate color--primary']"):
        song_name.append(i.text)
    # Artist Name
    for i in driver.find_elements_by_xpath("//span[@class = 'chart-element__information__artist text--truncate color--secondary']"):
        artist_name.append(i.text)
    # Last Week Rank
    for i in driver.find_elements_by_xpath("//div[@class = 'chart-element__meta text--center color--secondary text--last']"):
        last_week_rank.append(i.text)
    # Peak Rank
    for i in driver.find_elements_by_xpath("//div[@class = 'chart-element__meta text--center color--secondary text--peak']"):
        peak_rank.append(i.text)
    # Weeks on Board
    for i in driver.find_elements_by_xpath("//div[@class = 'chart-element__meta text--center color--secondary text--week']"):
        weeks_on_board.append(i.text)
    
    # Creating the dataframe.
    songs_df = pd.DataFrame({"Song Name" : song_name,
                             "Artist Name" : artist_name,
                             "Last Week Rank" : last_week_rank,
                             "Peak Rank" : peak_rank,
                             "Weeks On Board" : weeks_on_board})
    return songs_df


# Calling the function.
billboard("https://www.billboard.com/")
```

##### 9. Scrape the details most watched tv series of all time from imdb.com.

```python
# Function Definition
def tv_shows(url):
    driver =  webdriver.Chrome(r"F:\sanss_data\FlipRobo\chromedriver.exe")
    driver.get(url)
    
    # Creating empty lists for the data to be extracted.
    name = []
    year_span = []
    genre = []
    runtime = []
    ratings = []
    votes = []
    
    # Extracting the required data.
    # TV Show Name
    for i in driver.find_elements_by_xpath("//h3[@class = 'lister-item-header']/a"):
        name.append(i.text)
    # Year Span
    for i in driver.find_elements_by_xpath("//span[@class = 'lister-item-year text-muted unbold']"):
        year_span.append(i.text)
    # Genre
    for i in driver.find_elements_by_xpath("//span[@class = 'genre']"):
        genre.append(i.text)
    # Runtime
    for i in driver.find_elements_by_xpath("//span[@class = 'runtime']"):
        runtime.append(i.text)
    # Ratings
    for i in driver.find_elements_by_xpath("//div[@class = 'ipl-rating-star small']/span[2]"):
        ratings.append(i.text)
    # Votes
    for i in driver.find_elements_by_xpath("//span[@name = 'nv']"):
        votes.append(i.text)
    # Creating the DataFrame of the extracted data.
    
    df_tv_shows = pd.DataFrame({"Name" : name,
                                "Year Span" : year_span,
                                "Genre" : genre,
                                "Runtime" : runtime,
                                "Ratings" : ratings,
                                "Votes" : votes})
    return df_tv_shows
    
    

# Calling the function
tv_shows("https://www.imdb.com/list/ls095964455/")
```
